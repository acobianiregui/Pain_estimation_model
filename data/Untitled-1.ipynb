```python

import pandas as pd
import matplotlib.pyplot as plt

file_path = r".\Filtered\S_03_filtered.csv"
df = pd.read_csv(file_path, sep=None, engine='python')  # automatically detects separator

for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='ignore')

num_cols = len(df.columns)
fig, axs = plt.subplots(num_cols, 1, figsize=(12, 3*num_cols), sharex=True)

if num_cols == 1:
    axs = [axs]  # make iterable if only one column

for i, col in enumerate(df.columns):
    axs[i].plot(df.Seconds, df[col], color='blue')
    axs[i].set_ylabel(col)
    axs[i].set_title(f"{col} vs Index")

axs[-1].set_xlabel("Index")
plt.tight_layout()
plt.show()
```
